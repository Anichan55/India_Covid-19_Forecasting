```python
india_df = df[df['location']=='India'].groupby('date')['total_cases','total_deaths','population'].sum()
india_df['days'] = list(range(1,len(india_df)+1))
india_df['rate'] = (india_df.total_cases-india_df.total_cases.shift(1))/india_df.total_cases
india_df['active_cases']=(india_df.total_cases*0.34)
```

```python
india_df.tail(5)
```

```python
from scipy.optimize import curve_fit
import pylab
from datetime import timedelta



def sigmoid(x, L ,x0, k, b):
    y = L / (1 + np.exp(-k*(x-x0)))+b
    return (y)


xdata = india_df.days
ydata = india_df.total_cases

x_deaths=india_df.days
y_deaths=india_df.total_deaths

p0 = [max(ydata),np.median(xdata),1,min(ydata)]

population=df[df['location']=='India'][df[df['location']=='India']['date']=='2020-07-14'].population

```

```python
popt, pcov = curve_fit(sigmoid, xdata, ydata,p0,method='dogbox')
```

```python
popt
```

```python
popt[1]
```

```python
x = np.linspace(-1, 400, 50)
y = sigmoid(x, popt[0], popt[1], popt[2],0)

x_data=x
y_data=y*0.03
```

```python
pylab.plot(xdata, ydata,'o', label='Actual Cases')
pylab.plot(x,y, label='Predicted Cases')
pylab.plot(x_data,y_data, label='Total Deaths')
pylab.ylim(-0.05, 25000000)
pylab.xlim(-0.05, 500)
pylab.legend(loc='best')
plt.xlabel('Days')
plt.ylabel('Confirmed Cases')
plt.title('India')
pylab.show()
```
